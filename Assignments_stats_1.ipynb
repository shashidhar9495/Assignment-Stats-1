```python
## problem 1 

def mod(lis):
    l1=set(lis)
    rep=0
    for i in l1:
        if lis.count(i)>rep:
            rep=lis.count(i)
            max_rep=i
    return max_rep        
x=[6,7,5,7,7,8,7,9,7,4,10,6,8,8,9,5,6,4,8]
mean=np.mean(x)
mode=mod(x)
median=np.median(x)
std=np.std(x)
print('mean={}, mode={}, median={} and standard_deviation={}'.format(mean,mode,median,std))
```

```python
def fact(n):
    k=1
    for i in range(1,n+1):
        k=i*k
    return k
def binomial(n,x,p):
    factorial=fact(n)/(fact(n-x)*fact(x))
    comb=factorial*(p**x)*((1-p)**(n-x))
    return round(comb,4)
    
```

### problem 6

```python
import math
def poissons(mean,n):
    val=(math.exp(-mean))*(mean**n)/fact(n)
    return val
```

### problem7

given,
customers ariving per hour = 72

mean of customers ariving in 4 minutes = (72*4)/60


```python

mean=72*4/60
p_5=round(poissons(mean,5),4)
print('The probability of 5 customers arriving in 4 minutes is: {}'.format(p_5))

k=0
for i in range(4):
    k+=round(poissons(mean,i),4)
print('The probability of not more then 3 customers arriving in 4 minutes is: {}'.format(k))
print('The probability more then 3 customers arriving in 4 minutes is: {}'.format(1-k))

label=['probability <= 3 customers ariving in 4 min','probability > 3 customers ariving in 4 min']
plt.pie([k,1-k],labels=label,autopct='%1.2f%%')
plt.show()
```

### Problem 8
Given,

number of words per minute=77

errors per hour=6

number of words per hour=77*60=4620

avg error per 455 pages = (6*455)/4260 =  0.6408

```python
mean=0.6408
p_2=round(poissons(mean,2),4)
print('The probability that I will commit 2 errors in a 455-word financial report is: {}'.format(p_2))

mean_1=(6*1000)/4620
p_21=round(poissons(mean_1,2),4)
print('The probability that I will commit 2 errors in a 1000-word financial report is: {}'.format(p_21))

mean_2=(6*255)/4620
p_22=round(poissons(mean_2,2),4)
print('The probability that I will commit 2 errors in a 255-word financial report is: {}'.format(p_22))

label=['probability of error in a 455-word report','probability of error in a 1000-word report','probability of error in a 255-word report']
plt.pie([p_2,p_21,p_22],labels=label,autopct='%1.2f%%')
plt.show()
```

As the number of words in financial report increases the  value increases and probability of commiting error also increases.
